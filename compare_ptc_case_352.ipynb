This IPython notebook converts the OpenPTV results into `ptc*.dat` files following http://www.piv.jp/image3d/ptc3d.html

```python
import numpy as np
import matplotlib.pyplot as plt
import os
from glob import glob
%matplotlib inline

```

```python
ptv_path = '/Users/alex/Dropbox/Testfor352/ptc/'
piv_path = '/Users/alex/Dropbox/352/piv_std_352/ptc/'

ptc_openptv = glob(ptv_path+'ptc*.dat')
ptc_3dpiv = glob(piv_path+'ptc*.dat')

print ptc_openptv[0]
print ptc_3dpiv[0]
```

```python
dtype = {'names':('id','x','y','z','x1','y1','x2','y2','x3','y3','intensity'),'formats':('i4','f','f','f','f','f','f','f','f','f','f','f','f')}
ptc1 = np.loadtxt(ptc_openptv[120],dtype=dtype)
ptc2 = np.loadtxt(ptc_3dpiv[120],dtype=dtype)
```

```python
# find those which are in the field of view in all three cameras
id1 = ptc2['x1'] == 0.0
id2 = ptc2['x2'] == 0.0
id3 = ptc2['x3'] == 0.0
nid = ~np.logical_or(id1,id2,id3)
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig = plt.figure(figsize=(16,12))
ax = fig.add_subplot(1,1,1, projection='3d')

x = ptc2['x']
y = ptc2['y']
z = ptc2['z']
ax.scatter(x,y,z,zdir='y',c='w',marker='o',s=50,alpha=0.5)

ax.hold(True)

x1 = ptc1['x']
y1 = ptc1['y']
z1a = ptc1['z']

zshift = 0.043
z1 = z1a-zshift


ax.scatter(x1,y1,z1,zdir='y',marker='x',c='r',s=50)
ax.view_init(elev=10., azim=25.)
ax.set_xlabel('$x$',fontsize=18)
ax.set_ylabel('$y$',fontsize=18)
ax.set_zlabel('$z$',fontsize=18)
```

```python
from scipy.spatial import cKDTree
```

```python
tree = cKDTree(zip(x, y, z))
d, inds = tree.query(zip(x1, y1, z1), k = 1)
```

```python
print x1[0],y1[0],z1[0]
print x[inds[0]],y[inds[0]],z[inds[0]]
```

```python
from __future__ import division
print "percentage of identified points %3.1f " % (100*len(inds)/sum(nid))
```

```python
print "average distance %3.2f mm" % (sum(d)/len(inds)*10)
```

```python
plt.figure(figsize=(15,10))
plt.plot(x[nid],y[nid],'wo',x1,y1,'rx',markersize=10)
plt.xlabel('$x$ [cm]',fontsize=18),plt.ylabel('$y$ [cm]',fontsize=18)
plt.title('Top view')
```

```python
plt.figure(figsize=(15,10))
plt.plot(x[nid],z[nid],'wo',x1,z1,'rx',markersize=10)
plt.xlabel('$x$ [cm]',fontsize=18),plt.ylabel('$z$ [cm]',fontsize=18)
plt.title('Side view')
```
