```python
from matplotlib import pyplot
import os, sys
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import animation

# sys.path.insert(0,os.path.abspath('/Users/alex/Documents/OpenPTV/postptv/'))
from flowtracks.io import trajectories_ptvis
%matplotlib inline
```

```python
inName = '/Users/alex/Dropbox/Testfor352/res/ptv_is.%d'
trajects = trajectories_ptvis(inName, traj_min_len=5) # the directory with the input files
```

```python
fig = pyplot.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1, projection='3d')
positions = [tr.pos() for tr in trajects]    


def init():
    for pos in positions[:-1]:
        trajX = [pos[k][0] for k in range(len(pos))]
        trajY = [pos[k][1] for k in range(len(pos))]
        trajZ = [pos[k][2] for k in range(len(pos))]
        ax.plot(trajX, trajZ, trajY)

def animate(i):
    ax.view_init(elev=10., azim=i*5)

# Animate
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=36, interval=20, blit=True)


# Save
anim.save('basic_animation.mp4', fps=30, extra_args=['-vcodec', 'libx264'])
```
