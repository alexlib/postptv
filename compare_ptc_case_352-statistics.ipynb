This IPython notebook converts the OpenPTV results into `ptc*.dat` files following http://www.piv.jp/image3d/ptc3d.html

```python
import numpy as np
import matplotlib.pyplot as plt
import os
from glob import glob
from scipy.spatial import cKDTree
from __future__ import division
%matplotlib inline
```

```python
ptv_path = '/Users/alex/Dropbox/Testfor352/ptc/'
piv_path = '/Users/alex/Dropbox/352/piv_std_352/ptc/'

ptc_openptv = glob(ptv_path+'ptc*.dat')
ptc_3dpiv = glob(piv_path+'ptc*.dat')

print ptc_openptv[0]
print ptc_3dpiv[0]
```

```python
dtype = {'names':('id','x','y','z','x1','y1','x2','y2','x3','y3','intensity'),'formats':('i4','f','f','f','f','f','f','f','f','f','f','f','f')}
```

```python
ave_dist, percentage = [],[]
zshift = 0.043

for i in range(len(ptc_openptv)):
    ptc1 = np.loadtxt(ptc_openptv[i],dtype=dtype)
    ptc2 = np.loadtxt(ptc_3dpiv[i],dtype=dtype)

    # find those which are in the field of view in all three cameras
    id1 = ptc2['x1'] == 0.0
    id2 = ptc2['x2'] == 0.0
    id3 = ptc2['x3'] == 0.0
    nid = ~np.logical_or(id1,id2,id3)

    x = ptc2['x']
    y = ptc2['y']
    z = ptc2['z']

    x1 = ptc1['x']
    y1 = ptc1['y']
    z1 = ptc1['z'] - zshift

    tree = cKDTree(zip(x, y, z))
    d, inds = tree.query(zip(x1, y1, z1), k = 1)

    percentage.append(len(inds)/sum(nid))
    ave_dist.append(sum(d)/len(inds)*10)
```

```python
plt.figure(figsize=(12,8))
plt.plot(percentage,'o:')
plt.xlabel('frames',fontsize=16)
plt.ylabel('success ratio',fontsize=16)
```

```python
plt.figure(figsize=(12,8))
plt.plot(ave_dist,'o:')
plt.xlabel('frames',fontsize=16)
plt.ylabel('average distance [mm]',fontsize=16)
```
